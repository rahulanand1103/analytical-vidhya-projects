## 1.import library

```
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score
data=pd.read_csv("adult.csv")
```

```
data
```

```
data[data == '?'] = np.nan
```

```
print(data.isnull().sum(axis = 0))
```

## 2.handling missing values

```
for col in ['workclass', 'occupation', 'native.country']:
    data[col].fillna(data[col].mode()[0], inplace=True)
```

```
data
```

## 3.One-hot encoding

```
#One Hot Coding:
data = pd.get_dummies(data, columns=['workclass','education','marital.status','occupation',
                              'relationship','race','sex','native.country'])
```

```
data
```

## 4.splitting data into test and train

```
income=data['income'].value_counts()
print(income)
```

```
data['income']=data['income'].map({'<=50K':1,'>50K':0})
```

```
y = data['income']
data.drop(['income','fnlwgt'], axis=1,inplace=True)
```

```
train,test,y_train,y_test=train_test_split(data,y, test_size=0.2,shuffle=True)
```

```
print(train.shape,test.shape,y_train.shape,y_test.shape)
```

5.Logistic regression

```
LR= LogisticRegression()
LR.fit(train,y_train)
y_pred =LR.predict(test)
print("Accuracy on test set: %0.3f%%"%(accuracy_score(y_test, y_pred)*100))
```
