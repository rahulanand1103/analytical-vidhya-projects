```
import pandas as pd
import numpy as np
from sklearn.preprocessing import OneHotEncoder
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
from xgboost.sklearn import XGBRegressor
```

```
test=pd.read_csv("test.csv")
train=pd.read_csv("train.csv")
```

```
train.head()
```

```
print(train.shape,test.shape)
```

```
train=train.drop(["User_ID","Product_ID"],axis=1)
test=test.drop(["User_ID","Product_ID"],axis=1)
```

```
print("Train")
print(train.isnull().sum(axis = 0))
print("Test")
print(test.isnull().sum(axis = 0))
```

```
train['Product_Category_1'].fillna((train['Product_Category_1'].mean()), inplace=True)
train['Product_Category_2'].fillna((train['Product_Category_2'].mean()), inplace=True)
train['Product_Category_3'].fillna((train['Product_Category_3'].mean()), inplace=True)

test['Product_Category_2'].fillna((test['Product_Category_2'].mean()), inplace=True)
test['Product_Category_3'].fillna((test['Product_Category_3'].mean()), inplace=True)
```

```
train['Gender']=train['Gender'].map({'F':1,'M':0})
test['Gender']=test['Gender'].map({'F':1,'M':0})
```

```
train
```

```
train['train']=1
test['train']=0
encoding=pd.concat([train,test])
```

```
#One Hot Coding:
encoding = pd.get_dummies(encoding, columns=['City_Category','Age','Stay_In_Current_City_Years'])
encoding
```

```
train=encoding[encoding['train']==1]
test=encoding[encoding['train']==0]
```

```
y_train=train.Purchase
train.drop(['Purchase'],axis=1,inplace=True)
```

```
train
```

```
model = RandomForestRegressor()
model.fit(train, y_train)
y_pred = model.predict(test)
print(y_pred)
```
